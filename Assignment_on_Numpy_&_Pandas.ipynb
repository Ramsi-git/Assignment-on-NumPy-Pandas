#Q1. Employee Salary Adjustment

```python
#import numpy
import numpy as np
#A company stores salaries in a Numpy array.
salaries = np.array([45000, 52000, 61000, 48000, 75000])
```

```python
#Increase all salaries by 10%.
increased_salaries=salaries*1.10
print(increased_salaries)
```

```python
#Find the average salary.
avg_salary=np.mean(increased_salaries)
print(avg_salary)
```

```python
#Find the maximum and minimum salary.
max_sal=np.max(increased_salaries)
min_sal=np.min(increased_salaries)
print(max_sal)
print(min_sal)
```

```python
#Print salaries greater than 60000.
above_60k=increased_salaries[increased_salaries>60000]
print(above_60k)
```

#Q2. Sales Data Analysis

```python
#Daily sales of a store for 10 days:
sales = np.array([120,150,180,200,170,160,210,220,190,175])
```

```python
#Compute total sales.
total_sales = np.sum(sales)
print(total_sales)
```

```python
#Compute mean and standard deviation.
mean_sales = np.mean(sales)
std_sales = np.std(sales)
print(mean_sales)
print(std_sales)
```

```python
#Identify days where sales exceeded the average.
high_sales_days = sales[sales > mean_sales]
print(high_sales_days)
```

```python
#Normalize the sales using Min-Max scaling.
norm_sales = (sales - np.min(sales)) / (np.max(sales) - np.min(sales))
print(f"Normalized Sales: {norm_sales}")
```

#Q3. Matrix Operations for Business Data

```python
#A company has sales data for 3 products across 4 regions.
sales_matrix = np.array([[120,140,130,150], [80,90,100,95], [200,210,190,205]])
```

```python
#Find total sales per product.
prod_totals = np.sum(sales_matrix, axis=1)
print(prod_totals)
```

```python
#Find total sales per region.
region_totals = np.sum(sales_matrix, axis=0)
print(region_totals)
```

```python
#Identify the best performing product.
best_prod_idx = np.argmax(prod_totals) + 1
print(best_prod_idx)
```

```python
#Identify the region with highest sales.
best_region_idx = np.argmax(region_totals) + 1
print(best_region_idx)
```

#Q4. Employee Dataset Exploration

```python
#import pandas
import pandas as pd
#Create a DataFrame:
data = {
'Employee':['A','B','C','D','E','F'],
'Department':['HR','IT','IT','Finance','HR','Finance'],
'Salary':[50000,65000,70000,72000,52000,68000],
'Experience':[2,5,7,6,3,4]
}
```

```python
#Convert this dictionary into a DataFrame.
df = pd.DataFrame(data)
print(df)
```

```python
#Display basic statistics of the dataset.
stats = df.describe()
print(stats)
```

```python
#Find the average salary per department.
dept_avg = df.groupby('Department')['Salary'].mean()
print(dept_avg)
```

```python
#Identify employees with experience greater than 4 years.
senior_emp = df[df['Experience'] > 4]
print(senior_emp)
```

```python
#Add a new column Bonus = 10% of salary.
df['Bonus'] = df['Salary'] * 0.10
print(df)
```

#Q5. Missing Data Handling

```python
#Dataset:
data = {
'Name':['John','Anna','Peter','Linda','David'],
'Age':[28, np.nan, 35, 32, np.nan],
'Salary':[50000,54000,np.nan,58000,60000]
}
df_missing = pd.DataFrame(data)
print(df_missing)
```

```python
#Detect missing values.
print(df_missing.isnull().sum())
```

```python
#Replace missing Age with mean age.
df_missing['Age'] = df_missing['Age'].fillna(df_missing['Age'].mean())
print(df_missing)
```

```python
#Replace missing Salary with median salary.
df_missing['Salary'] = df_missing['Salary'].fillna(df_missing['Salary'].median())
print(df_missing)
```

```python
#Remove rows with any remaining missing values.
df_cleaned = df_missing.dropna()
print(df_cleaned)
```

#Q6. Sales Performance Analysis

```python
#Dataset:
sales_data = {
'Region':['North','South','East','West','North','South','East','West'],
'Salesperson':['A','B','C','D','E','F','G','H'],
'Sales':[12000,15000,13000,17000,16000,14000,15500,16500]
}
df_sales = pd.DataFrame(sales_data)
print(df_sales)
```

```python
#Find total sales per region.
region_sales = df_sales.groupby('Region')['Sales'].sum()
print(region_sales)
```

```python
#Identify the top salesperson.
top_person = df_sales.loc[df_sales['Sales'].idxmax()]
print(top_person)
```

```python
#Compute average sales across regions.
avg_reg_sales = df_sales.groupby('Region')['Sales'].mean()
print(avg_reg_sales)
```

```python
#Sort the dataset by sales descending.
df_sorted = df_sales.sort_values(by='Sales', ascending=False)
print(df_sorted)
```

#Q7. Customer Purchase Analysis

```python
#Dataset:
data_cust= {
'CustomerID':[101,102,103,104,105,106,107],
'Age':[25,45,35,50,23,40,60],
'PurchaseAmount':[250,600,400,800,150,700,900],
'City':['Delhi','Mumbai','Delhi','Chennai','Mumbai','Delhi','Chennai']
}
df_cust = pd.DataFrame(data_cust)
print(df_cust)
```

```python
#Find the average purchase amount by city
city_avg = df_cust.groupby('City')['PurchaseAmount'].mean()
print(city_avg)
```

```python
# Identify customers with purchase above overall average.
high_buyers = df_cust[df_cust['PurchaseAmount'] > df_cust['PurchaseAmount'].mean()]
print(high_buyers)
```

```python
#Find the city generating the highest revenue.
top_city = df_cust.groupby('City')['PurchaseAmount'].sum().idxmax()
print("Top City:" ,top_city)
```

```python
#Create a column Age_Group:
# <30 → Young
# 30–50 → Adult
# 50 → Senior
def age_cat(age):
    if age < 30: return 'Young'
    elif 30 <= age <= 50: return 'Adult'
    else: return 'Senior'

df_cust['Age_Group'] = df_cust['Age'].apply(age_cat)
print(df_cust)
```

```python
#Find average purchase per age group.
age_group_avg = df_cust.groupby('Age_Group')['PurchaseAmount'].mean()
print(age_group_avg)
```

#Q8: Time Series Sales Analysis

```python
#Generate data:
dates = pd.date_range('2024-01-01', periods=12, freq='ME') # 'ME' for Month End
sales_ts = [200, 220, 250, 270, 300, 320, 350, 360, 380, 400, 420, 450]
df_ts = pd.DataFrame({'Date': dates, 'Sales': sales_ts})
print(df_ts)
```

```python
#Convert Date to datetime index.
df_ts.set_index('Date', inplace=True)
print(df_ts.index)
```

```python
#Compute monthly percentage growth.
df_ts['Growth'] = df_ts['Sales'].pct_change() * 100
print(df_ts[['Sales','Growth']])
```

```python
#Calculate 3-month rolling average sales.
df_ts['Rolling_Avg'] = df_ts['Sales'].rolling(window=3).mean()
print(df_ts[['Sales','Rolling_Avg']])
```

```python
#Identify the month with highest growth rate.
highest_growth_month = df_ts['Growth'].idxmax()
print(df_ts)
print(f"Highest Growth Month: {highest_growth_month}")
```
